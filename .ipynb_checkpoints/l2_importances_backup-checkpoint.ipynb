### Addendum - Feature Importances

#### Derived from the Model

```python
# get feature importances
importances=model.feature_importances_

# get names of encoded features
encoded_feature_names=preprocessor.get_feature_names_out()

# create a DataFrame to display the importances
feature_importances_df = pd.DataFrame({
    'Feature': encoded_feature_names,
    'Importance': importances
})

# sort by importance
feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)

print('Top 10 Features')
print(feature_importances_df.head(10))  # Display top 10 features

# Plot sorted feature importances
feature_df_sorted = feature_importances_df.sort_values('Importance', ascending=True)
feature_df_sorted.plot.barh(y='Importance', x='Feature', figsize=(8, 8))
plt.title('Feature Importances from RandomForestClassifier')
plt.xlabel('Importance')
plt.show()
```

We can see that the most important features for the model were, in order from most important:
1. Call Duration
1. Balance
1. Day
1. Age
